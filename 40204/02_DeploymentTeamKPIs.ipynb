<div dir="rtl">
    <p style="text-align:right; font-family: Vazirmatn, B Nazanin, Tahoma; font-size:11pt; line-height:2; padding-right:10px;">
        <b>فرمول ریاضی</b> <br>
        <span style="text-align:right; font-family: Vazirmatn, B Nazanin, Tahoma; font-size:11pt; line-height:2; color: #656565; padding-right:20px;"> ((زمان استقرار قبلی - زمان استقرار هدف) / زمان استقرار هدف) * ۱۰۰ </span><br>
    </p>
</div>


$$Percent\ of\ Reduction\ =\ \frac{Previous\ Deployment_{time}\ -\ Target\ Deployment_{time}}{Previous\ Deployment_{time}}\ \times\ 100$$


<br><br>

<div dir="rtl">
    <p style="text-align:right; font-family: Vazirmatn, B Nazanin, Tahoma; font-size:11pt; line-height:2; padding-right:10px;">
        <b>کد پایتون:</b> <br>
    </p>
</div>
<br>

``` python
previous_deployment_time = 10  # زمان استقرار قبلی به طور مثال 10 روز کاری
target_deployment_time = 1  # زمان استقرار هدف به طور مثال 1 روز کاری

percentage_reduction = ((previous_deployment_time - target_deployment_time) / previous_deployment_time) * 100
percentage_reduction = round(percentage_reduction, 2)  # گرد کردن به دو رقم اعشاری

print(f"درصد کاهش زمان استقرار: {percentage_reduction}%")


```


<div dir="rtl">
    <p style="text-align:right; font-family: Vazirmatn, B Nazanin, Tahoma; font-size:11pt; line-height:2; padding-right:10px;">
        سوالات مرتبط با شاخص   
    </p>
    <span style="text-align:right; font-family: Vazirmatn, B Nazanin, Tahoma; font-size:11pt; line-height:2; color: #656565; padding-right:20px;"> ۱. نتیجه‌ای که می‌خواهید به‌دست آورید، چیست؟ &emsp;&emsp; کاهش زمان استقرار نرم‌افزار به یک روز کاری</span><br>
    <span style="text-align:right; font-family: Vazirmatn, B Nazanin, Tahoma; font-size:11pt; line-height:2; color: #656565; padding-right:20px;"> ۲. چرا این نتیجه مهم است؟ &emsp;&emsp; کاهش زمان استقرار نرم‌افزار منجر به افزایش سرعت و کارایی در تحول نسخه‌های جدید و بهبود تجربه کاربری مشتریان می‌شود</span><br>
    <span style="text-align:right; font-family: Vazirmatn, B Nazanin, Tahoma; font-size:11pt; line-height:2; color: #656565; padding-right:20px;"> ۳. چگونه می‌توانید بر این نتیجه تأثیر بگذارید؟ &emsp;&emsp; با بهبود فرآیند استقرار، استفاده از ابزارها و روش‌های خودکار سازی و هماهنگی بیشتر با تیم‌های توسعه و تست</span><br>
    <span style="text-align:right; font-family: Vazirmatn, B Nazanin, Tahoma; font-size:11pt; line-height:2; color: #656565; padding-right:20px;"> ۴. چه کسی مسئول نتیجه است؟ &emsp;&emsp; تیم استقرار نرم‌افزار</span><br>
    <span style="text-align:right; font-family: Vazirmatn, B Nazanin, Tahoma; font-size:11pt; line-height:2; color: #656565; padding-right:20px;"> ۵. چگونه می‌خواهید پیشرفت را با نتیجه گفته‌شده اندازه‌گیری کنید؟ &emsp;&emsp; با اندازه‌گیری تعداد روزهای مورد نیاز برای استقرار نرم‌افزار</span><br>
    <span style="text-align:right; font-family: Vazirmatn, B Nazanin, Tahoma; font-size:11pt; line-height:2; color: #656565; padding-right:20px;"> ۶. چگونه می‌دانید که به نتیجه رسیده‌اید؟ &emsp;&emsp; با رسیدن زمان استقرار نرم‌افزار به یک روز کاری یا کمتر</span><br>
    <span style="text-align:right; font-family: Vazirmatn, B Nazanin, Tahoma; font-size:11pt; line-height:2; color: #656565; padding-right:20px;"> ۷. هر چند وقت یک بار پیشرفت خود را با نتیجه بررسی می‌کنید؟ &emsp;&emsp; هر ماه یکبار</span><br>
</div>

<br><br>

<div dir="rtl">
    <p style="text-align:right; font-family: Vazirmatn, B Nazanin, Tahoma; font-size:11pt; line-height:2; padding-right:10px;">
        OKRهای مرتبط برای شاخص‌ها  
    </p>
    <span style="text-align:right; font-family: Vazirmatn, B Nazanin, Tahoma; font-size:11pt; line-height:2; color: #656565; padding-right:20px;"> <b> هدف: </b> بهبود روند استقرار نرم‌افزار </span><br>
    <span style="text-align:right; font-family: Vazirmatn, B Nazanin, Tahoma; font-size:11pt; line-height:2; color: #656565; padding-right:20px;"> 
        <b> اهداف گروهی: </b><br>
        <span style="padding-right:30px;">
         ۱. ایجاد فرآیند استقرار استاندارد و بهینه برای تیم
        </span><br>
        <span style="padding-right:30px;">
        ۲. استفاده از ابزارها و روش‌های خودکار برای استقرار نرم‌افزار
        </span><br>
        <span style="padding-right:30px;">
        ۳. هماهنگی بیشتر با تیم‌های توسعه و تست
        </span>
    </span><br>
    <span style="text-align:right; font-family: Vazirmatn, B Nazanin, Tahoma; font-size:11pt; line-height:2; color: #656565; padding-right:20px;">
        <b> نتایج کلیدی: </b> <br>
        <span style="padding-right:30px;">
        ۱. کاهش میانگین زمان استقرار به یک روز کاری
        </span><br>
        <span style="padding-right:30px;">
        ۲. افزایش کارایی و دقت در فرآیند استقرار
        </span><br>
    </span>
    <span style="text-align:right; font-family: Vazirmatn, B Nazanin, Tahoma; font-size:11pt; line-height:2; color: #656565; padding-right:20px;">
        <b> اقدامات: </b> <br>
        <span style="padding-right:30px;">
        ۱. بررسی و بهبود فرآیند استقرار فعلی
        </span><br>
        <span style="padding-right:30px;">
        ۲. آموزش تیم در مورد استفاده از ابزارها و روش‌های خودکار
        </span><br>
        <span style="padding-right:30px;">
        ۳. برگزاری جلسات هماهنگی با تیم‌های توسعه و تست
        </span>
    </span><br>
</div>


<p style="text-align:center;">
    <img src="https://user-images.githubusercontent.com/85065799/204543278-26c507b6-400a-42e0-852f-2e09362f6e12.png" alt="Line" target="_blank">
</p>
